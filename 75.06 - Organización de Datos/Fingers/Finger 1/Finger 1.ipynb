```python
# Imports.
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Mostrar cuales son los 10 barrios cuyo valor de la propiedad promedio es mayor (tener en cuenta para esto el valor en dolares del m2).

```python
properties = pd.read_csv('Propiedades-0701.csv', low_memory = False)
neightboardhood = properties.loc[:, ['place_name', 'price_usd_per_m2', 'state_name']].dropna(inplace = False)
capitalProperties = neightboardhood.loc[neightboardhood.state_name.str.contains('Federal'),:]
capitalProperties = neightboardhood.loc[neightboardhood['state_name'] == 'Capital Federal',:]
capitalProperties = capitalProperties.loc[:, ['place_name', 'price_usd_per_m2']]
grouped = capitalProperties.groupby('place_name').mean().sort_values('price_usd_per_m2', ascending = False).head(10)
grouped
```

Realizar una visualización de los datos que permita mostrar el resultado obtenido.

```python
grouped.plot(kind='bar', rot=90, figsize=(10,8), color='red' ,fontsize=13); 
plt.title('Precio por m2 por barrio de Capital Federal', fontsize=20);
plt.xlabel('Barrios', fontsize=16);
plt.ylabel('Precio por metro cuadrado', fontsize=20);

#trips['start_day_of_week'].value_counts().plot(kind='bar', rot=0, figsize=(10,8), color='purple' ,fontsize=13);
#plt.title('Cantidad de viajes segun el dia', fontsize=20);
#plt.xlabel('Dia de la semana', fontsize=16);
#plt.ylabel('Cantidad de viajes', fontsize=20);
```
