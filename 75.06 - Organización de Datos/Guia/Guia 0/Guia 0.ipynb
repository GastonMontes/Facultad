```python
# Hago el import de los frameworks necesarios
import pandas as pd

pd.options.mode.chained_assignment = None
```

```python
# Leo la data.
crashes = pd.read_csv('apc.csv')
crashes
```

```python
#  a) Una de las fechas mas frecuentes es 31 de Agosto de 1988
dates = crashes[['Date']]
dates['count'] = 1
dates = dates.groupby(['Date'], as_index = False).count()
dates = dates.sort_values(by = 'count', ascending = False)
dates

# RTA: Verdadero.
```
