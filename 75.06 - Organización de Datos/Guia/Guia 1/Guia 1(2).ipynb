Luego de analizar los datos publicados en: https://www.kaggle.com/NUFORC/ufo-sightings Indicar cuál de las siguientes afirmaciones es correcta:	 
 		 a) 	El tercer estado con mas avistajes de ovnis es Texas.
 		 b) 	El promedio de duración de los ufos con formas cambiantes es superior al resto.
 		 c) 	El avistaje de menor duración se registro en el estado de Wyoming
 		 d) 	El avistaje de mayor duración en segundos se registro en Canadá.

```python
# a) El tercer estado con mas avistajes de ovnis es Texas.
UFOsDF = pd.read_csv('ufo-sightings/scrubbed.csv')

statesUFOs = UFOsDF.assign(count = 1).loc[:, ['state', 'count']]
statesUFOs.groupby('state').sum().sort_values('count', ascending = False)

# RTA: Falso - Texas es el 4to estado con mas avistajes.

# b) El promedio de duración de los ufos con formas cambiantes es superior al resto.
# RTA: Falso - Ver el notebook anterior.

# c) El avistaje de menor duración se registro en el estado de Wyoming (Es wy)
durationStatesUFOs = UFOsDF.loc[:, ['state', 'duration (seconds)']]
durationStatesUFOs.dropna(inplace = True)
durationStatesUFOs['duration (seconds)'] = durationStatesUFOs['duration (seconds)'].apply(pd.to_numeric, errors='coerse')
durationStatesUFOs.dropna(inplace = True)
durationStatesUFOs.sort_values('duration (seconds)', ascending = True)
# RTA: Verdadero - El avistaje mas corto fue en Wyoming (wy) de 0.001s.

# d) El avistaje de mayor duración en segundos se registro en Canadá.
durationCountryUFOs = UFOsDF.loc[:, ['country', 'duration (seconds)']]
durationCountryUFOs.dropna(inplace = True)
durationCountryUFOs['duration (seconds)'] = durationCountryUFOs['duration (seconds)'].apply(pd.to_numeric, errors='coerse')
durationCountryUFOs.dropna(inplace = True)
durationCountryUFOs.sort_values('duration (seconds)', ascending = False)
# RTA: Falso - El mayor avistaje se dio en Gran Bretaña y fue de 97836000.00s.
```
