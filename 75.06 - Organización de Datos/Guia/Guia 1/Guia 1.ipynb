Pregunta a resolver:
Luego de analizar los datos publicados en: https://www.kaggle.com/NUFORC/ufo-sightings Indicar cuál de las siguientes afirmaciones es correcta:	 
 		 a) 	La forma con mayor cantidad de registros en Los Angeles, CA es la forma de triangulo.
 		 b) 	Existen reportes de 7 países distintos.
 		 c) 	La forma hexagonal es una de las formas menos populares registradas.
 		 d) 	Los avistajes en forma de bola de fuego tienen un tiempo promedio de duración de aproximadamente el doble que el de los de forma de huevo.

```python
# Leo la data.
df = pd.read_csv('ufo-sightings/scrubbed.csv')
df.columns
```

```python
# Resuelvo el punto a)
# a) La forma con mayor cantidad de registros en Los Angeles, CA es la forma de triangulo.
# Agrego un count como columna.
cityShapeDF = df.assign(count = 1)

# Me quedo con las columnas que me importan.
cityShapeDF = df.loc[:,['state','city', 'shape', 'count']]

# Limpio los NAs.
cityShapeDF.dropna(inplace=True)

# Filtro por estado = ca y por ciudad = los angeles
cityShapeDF = cityShapeDF.loc[cityShapeDF.state.str.contains('ca'),:]

# Filtro por la ciudad "Los Angeles".
cityShapeDF = cityShapeDF.loc[cityShapeDF.city.str.contains('los angeles'),:]

# Me quedo solo con el shape y el count.
cityShapeDF = cityShapeDF.loc[:, ['shape', 'count']]
grouped = cityShapeDF.groupby('shape').sum().sort_values(('count'), ascending=False)
grouped

# RTA: Falso - La forma con mayor cantidad de registros en Los Angeles es la forma de Luz.
```

```python
# Resuelvo el punto b)
# b) Existen reportes de 7 países distintos.
countries = df.assign(count = 1).loc[:, ['country', 'count']]
countries.groupby(['country']).sum()

# RTA: Falso - Solo hay 5 paises con reportes.
```

```python
# Resuelvo el punto c)
# c) La forma hexagonal es una de las formas menos populares registradas.
forms = df.assign(count = 1).loc[:,['shape','count']]
forms.groupby(['shape']).sum().sort_values(('count'), ascending=False)

# RTA: Verdadero - La forma hexagonal solo fue vista 1 sola vez.
```

```python
# Resuelvo el punto d)
# d) Los avistajes en forma de bola de fuego tienen un tiempo promedio de duración de aproximadamente 
#    el doble que el de los de forma de huevo.
# Agrego un count como columna.
proms = df.loc[:,['duration (seconds)','shape']]

# Delete NaN.
proms.dropna(inplace=True)

# Convert duration to float.
proms['duration (seconds)'] = proms['duration (seconds)'].apply(pd.to_numeric, errors='coerse')

# Delte NaN again.
proms.dropna(inplace=True)

#proms = proms.groupby['shape'].sum
proms.groupby(['shape']).mean().sort_values(('duration (seconds)'), ascending=False)

# RTA: Falso - La forma bola de fuego tiene un promedio de 4023.941484s mientras que la forma de huevo tiene un promedio
#      de 1886.793281s lo que da una relacion de 2.13
```

6.  	Luego de analizar los datos publicados en: https://www.kaggle.com/NUFORC/ufo-sightings Indicar cuál de las siguientes afirmaciones es correcta:	 
 		 a) 	La forma mas común en los ufos registrados es circular
 		 b) 	El promedio de duración de los ufos con formas cambiantes es superior al resto.
 		 c) 	Gran Bretaña tiene mas reportes de avistajes que Canadá
 		 d) 	El país con avistajes de mayor duración promedio es Gran Bretaña

```python
# Resuelvo el punto a)
# a) La forma mas común en los ufos registrados es circular
shapeUFOs = df.assign(count = 1).loc[:,['shape', 'count']]
shapeUFOs.groupby('shape').sum().sort_values(('count'), ascending=False)

# Falso: la forma mas vista es la tipo luz('light').
```

```python
# Resuelvo el punto b)
# b) El promedio de duración de los ufos con formas cambiantes es superior al resto.
durationUFOs = df.loc[:,['shape', 'duration (seconds)']]

# Delete NaN.
durationUFOs.dropna(inplace=True)

# Convert duration to float.
durationUFOs['duration (seconds)'] = durationUFOs['duration (seconds)'].apply(pd.to_numeric, errors='coerse')

# Delte NaN again.
durationUFOs.dropna(inplace=True)

durationUFOs.groupby('shape').mean().sort_values(('duration (seconds)'), ascending=False)

# RTA: Falso - El promedio de duracion mayor de un tipo de ovni es 82827.405063s de la forma de cono.
```

```python
# Resuelvo el punto c)
# c) Gran Bretaña tiene mas reportes de avistajes que Canadá
countryUFOs = df.assign(count = 1).loc[:,['country', 'count']]
countryUFOs.groupby('country').sum()

# Falso - Canada tiene 3000 reportes mientras que Gran Bretaña solo 1905.
```

```python
# Resuelvo el punto d)
# d) El país con avistajes de mayor duración promedio es Gran Bretaña
durationCountryUFOs = df.loc[:,['country', 'duration (seconds)']]

# Delete NaN.
durationCountryUFOs.dropna(inplace=True)

# Convert duration to float.
durationCountryUFOs['duration (seconds)'] = durationCountryUFOs['duration (seconds)'].apply(pd.to_numeric, errors='coerse')

# Delte NaN again.
durationCountryUFOs.dropna(inplace=True)

durationCountryUFOs.groupby('country').mean().sort_values(('duration (seconds)'), ascending=False)

# RTA: Verdadero - Gran Bretaña tiene un promedio de 66061.321207s por avistaje.
```
