#2.  Luego de analizar los datos publicados en: https://www.kaggle.com/NUFORC/ufo-sightings Indicar cuál de las siguientes afirmaciones es correcta:	 
###  a) Existen 963 ufos reportados con forma de chevron
### b) El ultimo registro de avistaje de un ufo en forma de luz fue en Mayo del 2014.
### c) Existen mas reportes de ufos en forma de cruz que de diamante.
### d) La forma con mayor cantidad de registros en Los Angeles, CA es la forma de triangulo.

```python
# a) Existen 963 ufos reportados con forma de chevron.
chevronUFOS = ufosDf[['shape']]
chevronUFOS.dropna(subset = ['shape'], inplace = True)
chevronUFOS = chevronUFOS.loc[chevronUFOS['shape'].str.contains('chevron'), :]
chevronUFOS.count()

# RTA: Falso - Son 952 los avistajes reportados con forma de chevron.
```

```python
#b) El ultimo registro de avistaje de un ufo en forma de luz fue en Mayo del 2014.
lightUFOS = ufosDf[['datetime', 'shape']]
lightUFOS = lightUFOS.loc[lightUFOS['shape'] == 'light', :]
lightUFOS['datetime2'] = pd.to_datetime(lightUFOS['datetime'], format='%m/%d/%Y %H:%M', errors='coerce')
lightUFOS.dropna(inplace = True)

lightUFOS['anio'] = lightUFOS['datetime2'].apply(lambda x: x.strftime('%Y'))
lightUFOS['mes'] = lightUFOS['datetime2'].apply(lambda x: x.strftime('%m'))
lightUFOS = lightUFOS.loc[lightUFOS['anio'] >= '2014', :]
lightUFOS = lightUFOS.loc[lightUFOS['mes'] >= '05', :]
lightUFOS

# RTA: VERDADERO - No hay avistajes posteriores a Mayo de 2014.
```

```python
# c) Existen mas reportes de ufos en forma de cruz que de diamante.
shapeUFOS = ufosDf[['shape']]
shapeUFOS['count'] = 1
shapeUFOS = shapeUFOS.groupby('shape', as_index = False).count()
shapeUFOS = shapeUFOS.sort_values(by = 'count', ascending = False)
shapeUFOS

# RTA: FALSO = Hay 233 en forma de cruz y 1178 en forma de diamante.
```

```python
# d) La forma con mayor cantidad de registros en Los Angeles, CA es la forma de triangulo.
angUFOS = ufosDf[['shape', 'city']]
angUFOS = angUFOS.loc[angUFOS['city'].str.contains('los angeles'), :]
del angUFOS['city']
angUFOS['count'] = 1
angUFOS = angUFOS.groupby('shape', as_index = False).count()
angUFOS = angUFOS.sort_values(by = 'count', ascending = False)
angUFOS

# RTA: FALSO - La forma mas vista en Los Angeles es la de luz.
```
